```python
import numpy as np
import matplotlib.pyplot as plt

print("Found")
print(f"NumPy version: {np.__version__}")

#Parameters 
mu= 0.15 #Assuming baseline 15% annual growth
sigma=0.30 #Assuming volatility
dt= 1/252 # One day (252 trading days in a year)

#random daily return

random_return= np.random.normal(mu * dt, sigma * np.sqrt(dt))

print(f"Random daily return: {random_return: .6f}")
print(f"As a percentage: {random_return * 100:.3f}%")
print(f"\nBreakdown:")
print(f"Expected daily drift (mu * dt): {mu * dt:.6f}")
print(f"Daily volatility (sigma * sqrt(dt)): {sigma * np.sqrt(dt):.6f}")

# Starting price
price_today = 100.0

# Apply the random return
price_tomorrow = price_today * (1 + random_return)

print(f"Today's price: ${price_today:.2f}")
print(f"Random return: {random_return * 100:.3f}%")
print(f"Tomorrow's price: ${price_tomorrow:.2f}")
print(f"Dollar change: ${price_tomorrow - price_today:.2f}")

#Simulating 5 days

S0 = 100.0
days=5
prices=[S0]

print(f"Starting price: ${50:.2f}\n")

for day in range(days):
    daily_return= np.random.normal(mu * dt, sigma * np.sqrt(dt))
    new_price=prices[-1] * (1 + daily_return)

    prices.append(new_price)

    print(f"Day {day + 1}:")
    print(f"  Return: {daily_return * 100:+.3f}%")
    print(f"  Price: ${prices[-2]:.2f} → ${new_price:.2f}")
    print()

print(f"Final price after 5 days: ${prices[-1]:.2f}")
print(f"Total return: {(prices[-1] / prices[0] - 1) * 100:.2f}%")

# Full year simulation
S0 = 100.0
days = 252
prices = [S0]

for day in range(days):
    daily_return = np.random.normal(mu * dt, sigma * np.sqrt(dt))
    new_price = prices[-1] * (1 + daily_return)
    prices.append(new_price)

# Plot it
plt.figure(figsize=(12, 6))
plt.plot(prices, linewidth=2)
plt.axhline(y=S0, color='red', linestyle='--', label='Starting Price', alpha=0.5)
plt.title('One Year of Simulated Stock Prices', fontsize=14, fontweight='bold')
plt.xlabel('Trading Days')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Starting: ${prices[0]:.2f}")
print(f"Ending: ${prices[-1]:.2f}")
print(f"Return: {(prices[-1] / prices[0] - 1) * 100:.2f}%")


```
